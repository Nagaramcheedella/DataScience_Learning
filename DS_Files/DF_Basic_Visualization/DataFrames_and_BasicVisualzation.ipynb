```python
# We can Visulaize the data in form of Graphs using seaborn library which we already imported as sns
# Distrubtion Plot or Histogram Plot
# sns.distplot(<datafrmae[colName]>)

visual1 = sns.distplot(stats['InternetUsers']) 
```

```python
# Same we can get using displot()

visual1 = sns.displot(stats['InternetUsers']) 
```

```python
visual1 = sns.displot(data=stats['InternetUsers'], bins = 40) 
```

```python
visual1 = sns.distplot(stats['InternetUsers'], bins = 40) 
```

```python
# Box Plot
# x, y : strings, optional: Input variables; these should be column names in data.
# data : DataFrame: Tidy ("long-form") dataframe where each column is a variable and each row is an observation.
# sns.boxplot(data = <df>, x = <colName for x axis>, y = <colName for y axis>)

visualbox1 = sns.boxplot(data = stats, x=stats.IncomeGroup, y=stats.BirthRate)
```

```python
# Another way

visualbox1 = sns.boxplot(data = stats, x="IncomeGroup", y="BirthRate")
```

```python
# Linear Plot
# x, y : strings, optional: Input variables; these should be column names in data.
# fit_reg : bool, optional : If True, estimate and plot a regression model relating the x and y variables.
# data : DataFrame: Tidy ("long-form") dataframe where each column is a variable and each row is an observation.
# hue: In seaborn, the hue parameter determines which column in the data frame should be used for colour encoding.

visulalinear = sns.lmplot(data = stats, x = "InternetUsers", y="BirthRate", fit_reg = False, hue = "IncomeGroup")
```

```python
visulalinear = sns.lmplot(data = stats, x = "InternetUsers", y="BirthRate", fit_reg = True, hue = "IncomeGroup")
```

```python
# Line Plot
# x, y : strings, optional: Input variables; these should be column names in data.
# data : DataFrame: Tidy ("long-form") dataframe where each column is a variable and each row is an observation.


visualline = sns.lineplot(data = stats, x = "InternetUsers", y="BirthRate")
```
