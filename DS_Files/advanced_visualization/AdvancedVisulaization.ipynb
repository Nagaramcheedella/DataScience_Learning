```python
# Importing requirerd packages

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Read csv file using pandas.read_csv

movies = pd.read_csv('/content/Movie-Ratings.csv')
```

```python
# Rename columns to avoid spaces 

movies.columns = ['Film', 'Genere', 'CriticRating', 'AudienceRating','Budget','Year']
```

```python
movies.columns
```

```python
# As getting % for year is not a reasonable. Let us convert data type 'int' to 'category data type'
# To convert into category we use astype()

movies.Film = movies.Film.astype('category')
movies.Year = movies.Year.astype('category')
```

```python
# Joint Plot


visualJoint1 = sns.jointplot(data=movies, x= 'CriticRating', y='AudienceRating')
```

```python
# Stacked Histogram with simplified code
# We will get each category in a Genere and append that in a List
# let us use for loop 

list1 = list() # empty list to get unique geners
mylabel = list()

for gen  in movies.Genere.unique():
  list1.append(movies[movies.Genere == gen].Budget)
  mylabel.append(gen)

plt.hist(list1, bins = 30, rwidth = 1, label = mylabel,stacked = True)

plt.legend()
#plt.show()


```

```python
# Sub PLots

f, axes = plt.subplots(1,2,figsize = (10,5))
kd1 = sns.kdeplot(movies.Budget, movies.CriticRating, ax = axes[0])
kd2 = sns.kdeplot(movies.Budget, movies.AudienceRating, ax = axes[1])
```

```python
f, axes = plt.subplots(2,2,figsize = (20,10))
kd1 = sns.kdeplot(movies.Budget, movies.CriticRating, ax = axes[0,0],cmap='Blues')
kd2 = sns.kdeplot(movies.Budget, movies.AudienceRating, ax = axes[0,1],cmap='Blues')
kd3 = sns.kdeplot(movies.AudienceRating, movies.CriticRating, ax = axes[1,0],cmap='Blues')
kd4 = sns.kdeplot(movies.CriticRating, movies.AudienceRating, ax = axes[1,1],cmap='Blues')
```

```python
# add axes limits and some keywords
c = sns.FacetGrid(movies, row = 'Genere', col= 'Year', hue = 'Genere')
kws = dict(s=50, linewidth = 0.5,edgecolor = 'Black')
c= c.map(plt.scatter, 'CriticRating','AudienceRating',**kws)
c.set(xlim=(0,100),ylim=(0,100))
for ax in c.axes.flat:
  ax.plot((0,100),(0,100), c= 'Red',ls = '--')
plt.show()
```
