```python
#Importing required Libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# importing dataset

dataset = pd.read_csv("/content/Social_Network_Ads.csv")
```

```python
dataset.head()
```

```python
dataset.info()
```

**Independent and Dependent Variables**

```python
# Let us denote Independent variable with X
# .values is to get values
# Here in given dataset Age and Estimated Salary are Independent Columns
X = dataset.iloc[:, [2,3]].values

# Let us denote Dependent variable with y
# -1 indicates last column(Dependent column)
# Purchased are Dependent variables
y = dataset.iloc[:,4].values

#y = dataset.iloc[:,4].values we can also use 4 instaed of -1 if we know Dependent Col number
```

```python
X
```

```python
y
```

**Splitting Data Set into Train and Test Data**

```python
# Spliting data as Test data and Train Data

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state = 0)
```

```python
X_train
```

```python
y_train
```

```python
X_test
```

```python
y_test
```

**Feature Scaling**

```python
# Feature Scaling
# We generally use fit_transform for train data and transform for test data

from sklearn.preprocessing import StandardScaler

sc_X = StandardScaler()
X_train = sc_X.fit_transform(X_train)
X_test = sc_X.transform(X_test)
```

```python
X_test
```

### Machine Learing Algorithm - SUPPORT VECTOR MACHINES

```python
# Fitting the Model
# n_estimators: By default it is 10.
from sklearn.svm import SVC

classifier = SVC(kernel = 'rbf', random_state=0)
classifier.fit(X_train, y_train)
```

```python
# Predict my Test results

y_pred = classifier.predict(X_test)
```

```python
y_pred
```

```python
y_test
```

```python
# Making Confusion Matrix

from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
```

```python
cm
```

Accuracy = (66 + 24)/100 = 90

### Model is 90% Accurate

# VISUALISATION

```python
# Visualising the Test set results
from matplotlib.colors import ListedColormap
X_set, y_set = X_test, y_test
X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 1, stop = X_set[:, 0].max() + 1, step = 0.01),
                     np.arange(start = X_set[:, 1].min() - 1, stop = X_set[:, 1].max() + 1, step = 0.01))
plt.contourf(X1, X2, classifier.predict(np.array([X1.ravel(), X2.ravel()]).T).reshape(X1.shape),
             alpha = 0.75, cmap = ListedColormap(('red', 'green')))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1],
                c = ListedColormap(('red', 'green'))(i), label = j)
plt.title('Decision Tree (Test set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()
```

```python
# Visualising the Training set results
from matplotlib.colors import ListedColormap
X_set, y_set = X_train, y_train
X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 1, stop = X_set[:, 0].max() + 1, step = 0.01),
                     np.arange(start = X_set[:, 1].min() - 1, stop = X_set[:, 1].max() + 1, step = 0.01))
plt.contourf(X1, X2, classifier.predict(np.array([X1.ravel(), X2.ravel()]).T).reshape(X1.shape),
             alpha = 0.75, cmap = ListedColormap(('red', 'green')))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1],
                c = ListedColormap(('red', 'green'))(i), label = j)
plt.title('Decision Tree (Training set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()
```
